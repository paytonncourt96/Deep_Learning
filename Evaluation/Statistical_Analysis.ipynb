#Statistical Result Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import wilcoxon
```

##Histogram to check for Normality

```python
data = pd.read_csv(path)

sample1 = data['Sample 1']
sample2 = data['Sample 2']


plt.figure(figsize=(10, 6))
plt.hist(sample1, bins=10, alpha=0.7, label='Sample 1', edgecolor='black')



plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Model 1: Histogram of Sample Results')
plt.legend()


plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(sample2, bins=10, alpha=0.7, label='Sample 2', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Model 2: Histogram of Sample Results')
plt.legend()


plt.show()
```

## Mean, Median and Standard Deviation

```python
mean_sample1 = sample1.mean()
median_model1 = np.median(sample1)
std_sample1 = sample1.std(ddof=1)  #ddof=1 for sample SD

mean_sample2 = sample2.mean()
median_model2 = np.median(sample2)
std_sample2 = sample2.std(ddof=1)

print(f"Model 1: Mean = {mean_sample1:.2f}")
print(f"Model 1: Median = {median_model1}")
print(f"Model 1: Sample Standard Deviation = {std_sample1:.2f}")
print(f"Model 2: Mean = {mean_sample2:.2f}")
print(f"Model 2: Median {median_model2}")
print(f"Model 2: Sample Standard Deviation = {std_sample2:.2f}")
```

```python
print(f"Model 1: Median = {median_model1}")
print(f"Model 2: Median {median_model2}")
```

## Wilcoxon - Signed Ranked test

```python
stat, p = wilcoxon(sample1, sample2)

print(f"Wilcoxon Test Statistic = {stat}, p-value = {p}")
if p < 0.05:
    print("Significant difference between the two models.")
else:
    print("No significant difference between the two models.")
```
